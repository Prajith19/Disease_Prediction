```python
#defining a function' leap_year'
#example:-
#           input:- 2000
#           2000 is divisible by 4
#                is divisible by 100
#                is divisible by 400 
#return:- True
#        all the three conditions are TRUE so it's a leap_year
#        if either of one condition is FALSE it's not a leap_year
def is_leap_year(year):
    if year % 4 == 0:
        if year % 100 == 0:
            if year % 400 == 0:
                return True
            else:
                return False
        else:
            return True
    else:
        return False
```

```python
#defining the dependent and independent columns
#
#from the decision(decision.csv) data set
#for 'ind' columns remove the admit_date column
ind = desicion.drop('admit_date',axis='columns')
#from the 'ind' data set
#for 'indep' columns remove the highest column
#
#now the independent columns contains flu, covid, malaria, others(columns)
indep = ind.drop('highest',axis='columns')
#from the decision(decision.csv) data set
#for 'd' columns remove the admit_date column
d = desicion.drop('admit_date',axis='columns')  
#from the 'd' data set
#for 'de' columns remove the flu column
de = d.drop('flu',axis='columns')
#from the 'de' data set
#for 'dep' columns remove the covid column
dep = de.drop('covid',axis='columns')
#from the 'dep' data set
#for 'depe' columns remove the malaria column
depe = dep.drop('malaria',axis='columns')
#from the 'depe' data set
#for 'dept' columns remove the others column
#
#now the dependent column contains highest column
dept = depe.drop('others',axis='columns')

#Python script that uses the scikit-learn library to perform decision tree classification
from sklearn import tree
#A decision tree classifier is then created named 'model' and fit to the training data 'indep' and 'dept'
model=tree.DecisionTreeClassifier()
#The fit method trains the model
col = indep.columns
indepen = indep[col].values
model.fit(indepen,dept)
#score method calculates the accuracy of the model on the same training data.
model.score(indepen,dept)
print("............................................................................")
#printing the data inside the summa list in a sequential order
print(f'Predicted Flu count:- {summa[0]}')
print(f'Predicted Covid count:- {summa[1]}')
print(f'Predicted Malaria count:- {summa[2]}')
print(f'Predicted Other Disease count:- {summa[3]}')
print("............................................................................")
#predictions using the predict method on a new data instance stored in 'summa' 
#and stores the predicted result in 'decision'
decision = model.predict([summa])

print("                ")
print("............................................................................")
#The result is then printed to the console, providing a prediction of which disease
#is expected to be more prevalent in the given 'word_month' and 'year' for a given 'facility_main'.
if decision == 1:
    print(f'Description:- Flu is predicted to be more in {word_month} {year} for/in {facility_main}')
elif decision == 2:
    print(f'Description:- Covid is predicted to be more in {word_month} {year} for/in {facility_main}')
elif decision == 3:
    print(f'Description:- Malaria is predicted to be more in {word_month} {year} for/in {facility_main}')
elif decision == 4:
    print(f'Description:- Other Diseases are predicted to be more in {word_month} {year} for/in {facility_main}')
print("............................................................................")
print("                     ")
    
```
